They both belong to the same class but show different data when we call them. In a similar way, let’s create a class that will take different data with different variables. This can be done with the help of a **constructor**.

```python
class datascience:
    
    def __init__(self ,phone_number , email_id):
        
        self.phone_number = phone_number
        self.email_id = email_id
               
    
    def return_student_detials(self):
        return self.phone_number, self.email_id 
        
```

Here **__init__** is constructor it can be pre-build or user defined.
**self** reference is always compulsory to tell that this belongs to this class

```python
disha = datascience()
```

This is showing an error because we have promised that two arguments will be passed whenever we try to create a variable of a class.

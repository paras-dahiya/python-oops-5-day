# Example of Polymorphism in case of class and object

```python
class data_science:
    def syllabus(self) : 
        print("this is my method for data science syllbaus " )
```

```python
class web_dev :
    def syllabus(self) : 
        print("this is my method for web dev " )
```

```python
def class_parcer(class_obj) : 
    for i in class_obj : 
        i.syllabus()
```

Now let's create objects

```python
obj_data_science= data_science()
obj_web_dev = web_dev()
```

Now we are keeping both of these variable inside the list

```python
class_ojb = [obj_data_science , obj_web_dev]
```

Now let's pass this variable into class_parser class

```python
class_parcer(class_ojb)
```

This function behaves differently depending on the context. When I try to pass a data science object, it can access data science methods. Similarly, when I pass a web development object, it accesses the relevant web development methods. This demonstrates that we can achieve polymorphism even in this scenario. Regardless of the specific object we pass, the function adapts and accesses methods from different classes. Thus, I achieve polymorphisma single entity with behavior that dynamically changes based on the object passed into my function.

This is how I can implement **polymorphism** in Pythons object-oriented programming (OOP) concept.

# Inheritance

Just as a child tries to inherit many traits from their parents and ancestors, such as nature and behavior, you will find the concept of inheritance in programming.

**In inheritance, a class can inherit the properties of another class. Essentially, we are trying to pass down properties from a higher-level class to a lower-level class.**

```python
class parent:
    def test_parent(self):
        print("This is my parent class")
```

We have just defined a method in the parent class. Now, we will create a child class.

```python
class child(parent):
    pass
```

Now let's create an object of it

```python
child_obj = child()
```

```python
child_obj.test_parent()
```

Here, the child class is trying to inherit the properties of the parent class. Any functions that were defined in the parent class will be available in the child class.
Even though test_parent was not a function of the child class, it was still able to access the parent class.

## Multi level inheritance

```python
class class1 :
    def test_class1(self) : 
        print("this is my class1 " )
```

```python
class class2(class1) : 
    def test_class2(self) : 
        print("this is my class2" )
```

```python
class class3(class2) : 
    def test_class3(self) : 
        print("this is my class3 ")
```

```python
obj_class3  = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```

```python
obj_class3.test_class3()
```

## Multi inheritance

```python
class class1:
    def test_class1(self) : 
        print("this is my class 1" )
```

```python
class class2 :
    def test_class2(self) : 
        print("this is my class 2")
```

```python
class class3 (class1 , class2) : 
    pass
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```
